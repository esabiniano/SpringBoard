# How Happy is the World?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# import data
h15 = pd.read_csv('data/2015.csv')
h16 = pd.read_csv('data/2016.csv')
h17 = pd.read_csv('data/2017.csv')
h18 = pd.read_csv('data/2018.csv')
h19 = pd.read_csv('data/2019.csv')

# add year column to identify each data
h15['Year'] = 2015 
h16['Year'] = 2016 
h17['Year'] = 2017 
h18['Year'] = 2018 
h19['Year'] = 2019

#combine all
all_df = h15.append([h16, h17, h18, h19])
all_df.to_csv('data/all_years.csv', index = False)
```

```python
# combine all to have a large data frame of the 5 years
print(all_df.info())
all_df.head()
```

## Happiness distribution Across Time

```python
plt.style.use('ggplot')

plt.figure(figsize=(14,10))
plt.hist(all_df[all_df['Year']==2015]["Happiness Score"], bins=30, alpha=0.5, label="2015", edgecolor = 'black')
#plt.hist(all_df[all_df['Year']==2016]["Happiness Score"], bins=30, alpha=0.5, label="2016", edgecolor = 'black')
#plt.hist(all_df[all_df['Year']==2017]["Happiness Score"], bins=30, alpha=0.5, label="2017", edgecolor = 'black')
#plt.hist(all_df[all_df['Year']==2018]["Happiness Score"], bins=30, alpha=0.5, label="2018", edgecolor = 'black')
plt.hist(all_df[all_df['Year']==2019]["Happiness Score"], bins=30, alpha=0.5, label="2019", edgecolor = 'black')


plt.xlabel("Happiness Score", size=14)
plt.title("World's Happiness Score (2015 - 2019)")
plt.legend(loc='upper right')
plt.savefig('plots/happiness_score_15_v_19.png')
```

```python
plt.style.use('ggplot')

plt.figure(figsize=(14,10))
plt.hist(all_df[all_df['Year']==2015]["Happiness Score"], bins=30, alpha=0.5, label="2015", edgecolor = 'black')
plt.hist(all_df[all_df['Year']==2016]["Happiness Score"], bins=30, alpha=0.5, label="2016", edgecolor = 'black')
plt.hist(all_df[all_df['Year']==2017]["Happiness Score"], bins=30, alpha=0.5, label="2017", edgecolor = 'black')
plt.hist(all_df[all_df['Year']==2018]["Happiness Score"], bins=30, alpha=0.5, label="2018", edgecolor = 'black')
plt.hist(all_df[all_df['Year']==2019]["Happiness Score"], bins=30, alpha=0.5, label="2019", edgecolor = 'black')


plt.xlabel("Happiness Score", size=14)
plt.title("World's Happiness Score (2015 - 2019)")
plt.legend(loc='upper right')

plt.savefig('plots/happiness_all_years.png')
```

```python
# List of five airlines to plot
year = list(range(2015, 2020))
plt.figure(figsize=(14,10))
# Iterate through the five airlines
for y in year:
    # Subset to the airline
    subset = all_df[all_df['Year'] == y]
    
    # Draw the density plot
    sns.distplot(subset['Happiness Score'], hist = False, kde = True,
                 kde_kws = {'shade': True, 'linewidth': 3},
                 label = y)
    
# Plot formatting
plt.legend(prop={'size': 16}, title = 'Year')
plt.title('Happiness Score Density Plot')
plt.xlabel('Happiness Score')
plt.ylabel('Density')

plt.savefig('plots/allyears_density_plt.png')
```

```python
# look at the correlation of happiness with the other variables
fig = plt.figure(figsize=(25,17))
fig.subplots_adjust(hspace=1)
ax1 = fig.add_subplot(321)
ax2 = fig.add_subplot(322)
ax3 = fig.add_subplot(323)
ax4 = fig.add_subplot(324)
ax5 = fig.add_subplot(325)
ax_l = [ax1,  ax2, ax3, ax4, ax5]

for y, a in zip(year, ax_l):
    # Subset to the airline
    subset = all_df[all_df['Year'] == y].drop(['Year', 'Lower Confidence Interval',
                                               'Upper Confidence Interval', 'Dystopia Residual',
                                              'Standard Error', 'Happiness Rank'], axis=1)
    corr = subset.corr()
    ax = sns.heatmap(corr, vmin=-1, vmax=1, center=0, cmap=sns.diverging_palette(20, 220, n=200), square=True, ax = a)
    ax.set_xticklabels(ax.get_xticklabels(), horizontalalignment='right')
    ax.set_title(y, fontsize = 16)
plt.savefig('plots/allyears_corr_plt.png')
```

```python
# look at the correlation of happiness with the other variables
plt.figure(figsize=(25,17))

subset = all_df[(all_df['Year'] == 2019) & (all_df['Region']=='North America')].drop(['Year', 'Lower Confidence Interval',
                                               'Upper Confidence Interval', 'Dystopia Residual',
                                              'Standard Error', 'Happiness Rank'], axis=1)
corr = subset.corr()
#ax = sns.heatmap(corr, vmin=-1, vmax=1, center=0, cmap=sns.diverging_palette(20, 220, n=200), square=True)
#ax.set_xticklabels(ax.get_xticklabels(), horizontalalignment='right')
#ax.set_title('2019 North America', fontsize = 16)
#plt.savefig('plots/2019_NAR_corr_plt.png')

mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(12, 12))
    ax = sns.heatmap(corr, mask=mask, vmax=1, square=True)
plt.savefig('plots/2019_NAR_corr_plt.png')
```
